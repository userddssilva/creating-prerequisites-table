# Import libres

```python
import pandas as pd
import numpy as np
from pprint import pprint
```

## Open grid of subjects

```python
grid = pd.read_excel(r'gride.xlsx')
```

```python
grid
```

## Rename the columns

```python
grid.columns = ['disciplina', 'pre_requisito', 'equivalencia']
grid.head()
```

# Separating the columns

## Get the codes of subjects

```python
code_sub = grid.disciplina.apply(lambda x : x.split(' - ')[0])
```

```python
code_sub
```

## Get the name of subjects

```python
name_sub = grid.disciplina.apply(lambda x : x.split(' - ')[1])
```

```python
name_sub
```

## Get the codes of prerequisites

```python
def separate(x):
    if type(x) == float:
        return ''
    else:
        return x.split(' E ')

prerequisites = grid.pre_requisito.apply(lambda x: separate(x))
```

```python
prerequisites
```

## Get the codes of equivalences

```python
def separator(x):
    if type(x) == float:
        return ''
    else:
        return x.split(' OU ')

equivalences = grid.equivalencia.apply(lambda x: separator(x))
```

```python
equivalences
```

# Creating new grid 

```python
new_grid = {
    'cod_disciplina': list(),
    'nome_disciplina': list(),
    'prerequisito': list(),
    'equivalencias': list(),
}
```

## Converting feartures to lists

```python
cod = list(code_sub)
name = list(name_sub)
pre = list(prerequisites)
equi = list(equivalences)
```

## Correcting the name of matter

```python
name[30] = 'Interação Homem - Computador'
```

```python
name
```

## Mounting new grid

```python
_index = 0
for list_pre, list_equi in zip(pre, equi):
    if list_pre == '' and list_equi == '':
        new_grid['cod_disciplina'].append(cod[_index])
        new_grid['nome_disciplina'].append(name[_index])
        new_grid['prerequisito'].append('')
        new_grid['equivalencias'].append('')
    elif list_pre == '' and list_equi != '':
        for equi_ in list_equi: 
            new_grid['cod_disciplina'].append(cod[_index])
            new_grid['nome_disciplina'].append(name[_index])
            new_grid['prerequisito'].append('')
            new_grid['equivalencias'].append(equi_)
    elif list_pre != '' and list_equi == '':
        for pre_ in list_pre:
            new_grid['cod_disciplina'].append(cod[_index])
            new_grid['nome_disciplina'].append(name[_index])
            new_grid['prerequisito'].append(pre_)
            new_grid['equivalencias'].append('')
    else:
        for pre_ in list_pre:
            for equi_ in list_equi:
                new_grid['cod_disciplina'].append(cod[_index])
                new_grid['nome_disciplina'].append(name[_index])
                new_grid['prerequisito'].append(pre_)
                new_grid['equivalencias'].append(equi_)
    _index += 1
```

```python
pprint(new_grid)
```

## Converting dict to DataFrame

```python
new_grid = pd.DataFrame.from_dict(new_grid)
```

```python
new_grid
```

## Exporting DataFrame

```python
new_grid.to_excel('new_grid.xlsx')
```
